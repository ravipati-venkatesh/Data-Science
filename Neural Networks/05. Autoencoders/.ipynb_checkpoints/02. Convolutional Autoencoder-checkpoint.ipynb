```python
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization,Conv2DTranspose
from keras.models import Model
from keras.datasets import mnist
import livelossplot
loss_plot = livelossplot.PlotLossesKeras()
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train/255
X_test = X_test/255
```

```python
def get_encoder_decoder_with_upsampling():
    # Encoder
    input_ = Input(shape=(28, 28, 1))
    encoder_conv1 = Conv2D(16, (3, 3), padding='same', activation='relu', name="encoder_conv1")(input_)
    encoder_batch_norm1 = BatchNormalization()(encoder_conv1)
    encoder_maxpooling1 = MaxPooling2D((2, 2), padding='same', name="encoder_maxpooling1")(encoder_batch_norm1)

    encoder_conv2 = Conv2D(8, (3, 3), padding='same', activation='relu', name="encoder_conv2")(encoder_maxpooling1)
    encoder_batch_norm2 = BatchNormalization()(encoder_conv2)
    encoder_maxpooling2 = MaxPooling2D((2, 2), padding='same', name="encoder_maxpooling2")(encoder_batch_norm2)

    encoder_conv3 = Conv2D(8, (3, 3), padding='same', activation='relu', name="encoder_conv3")(encoder_maxpooling2)
    encoder_batch_norm3 = BatchNormalization()(encoder_conv3)
    encoder_maxpooling3 = MaxPooling2D((2, 2), padding='same', name="encoder_maxpooling3")(encoder_batch_norm3)

    # Define encoder model
    encoder = Model(input_, encoder_maxpooling3)
#     encoder.summary()

    # Decoder
    decoder_input = Input(shape=(4, 4, 8))
    decoder_conv1 = Conv2D(8, (3, 3), padding='same', activation='relu', name="decoder_conv1")(decoder_input)
    decoder_batch_norm1 = BatchNormalization()(decoder_conv1)
    decoder_upsampling1 = UpSampling2D((2, 2), name="decoder_upsampling1")(decoder_batch_norm1)

    decoder_conv2 = Conv2D(8, (3, 3), padding='same', activation='relu', name="decoder_conv2")(decoder_upsampling1)
    decoder_batch_norm2 = BatchNormalization()(decoder_conv2)
    decoder_upsampling2 = UpSampling2D((2, 2), name="decoder_upsampling2")(decoder_batch_norm2)

    decoder_conv3 = Conv2D(16, (3, 3), activation='relu', name="decoder_conv3")(decoder_upsampling2)
    decoder_batch_norm3 = BatchNormalization()(decoder_conv3)
    decoder_upsampling3 = UpSampling2D((2, 2), name="decoder_upsampling3")(decoder_batch_norm3)

    decoder_conv4 = Conv2D(1, (3, 3), padding='same', activation='sigmoid', name="decoder_conv4")(decoder_upsampling3)

    # Define decoder model
    decoder = Model(decoder_input, decoder_conv4)
#     decoder.summary()

    # Full autoencoder model
    autoencoder_input = Input(shape=(28, 28, 1))
    encoded_output = encoder(autoencoder_input)
    decoded_output = decoder(encoded_output)
    autoencoder = Model(autoencoder_input, decoded_output)
#     autoencoder.summary()
    return encoder, decoder, autoencoder
```

```python
def get_encoder_decoder_with_conv_transpose():
    # Encoder
    encoder_input_ = Input(shape=(28, 28, 1))
    encoder_conv1 = Conv2D(16, (3, 3), padding='same', activation='relu', name="encoder_conv1")(encoder_input_)
    encoder_batch_norm1 = BatchNormalization()(encoder_conv1)
    encoder_maxpooling1 = MaxPooling2D((2, 2), padding='same', name="encoder_maxpooling1")(encoder_batch_norm1)

    encoder_conv2 = Conv2D(8, (3, 3), padding='same', activation='relu', name="encoder_conv2")(encoder_maxpooling1)
    encoder_batch_norm2 = BatchNormalization()(encoder_conv2)
    encoder_maxpooling2 = MaxPooling2D((2, 2), padding='same', name="encoder_maxpooling2")(encoder_batch_norm2)

    encoder_conv3 = Conv2D(8, (3, 3), padding='same', activation='relu', name="encoder_conv3")(encoder_maxpooling2)
    encoder_batch_norm3 = BatchNormalization()(encoder_conv3)

    # Define encoder model
    encoder = Model(encoder_input_, encoder_batch_norm3)
#     encoder.summary()

    # Decoder
    decoder_input = Input(shape=(7, 7, 8))
    conv_transpose1 = Conv2DTranspose(8, (3, 3), strides=2, activation="relu", padding="same")(decoder_input)
    decoder_batch_norm1 = BatchNormalization()(conv_transpose1)
    
    conv_transpose2 = Conv2DTranspose(8, (4, 4), strides=2, activation="relu", padding="same")(decoder_batch_norm1)
    decoder_batch_norm2 = BatchNormalization()(conv_transpose2)
    
#     conv_transpose3 = Conv2DTranspose(16, (3, 3), strides=2, activation="relu", padding="same")(decoder_batch_norm2)
#     decoder_batch_norm3 = BatchNormalization()(conv_transpose3)


    decoder_conv4 = Conv2D(1, (3, 3), padding='same', activation='sigmoid', name="decoder_conv4")(decoder_batch_norm2)

    # Define decoder model
    decoder = Model(decoder_input, decoder_conv4)
#     decoder.summary()

    # Full autoencoder model
    autoencoder_input = Input(shape=(28, 28, 1))
    encoded_output = encoder(autoencoder_input)
    decoded_output = decoder(encoded_output)
    autoencoder = Model(autoencoder_input, decoded_output)
#     autoencoder.summary()
    return encoder, decoder, autoencoder
```

```python
def compile_train(autoencoder, X_train, X_test, epoch=10):
    autoencoder.compile(optimizer='adam',
             loss='binary_crossentropy')

    autoencoder.fit(X_train, X_train,
                   epochs=epoch,
                   batch_size=1000,
                    callbacks=[loss_plot],
                   validation_data=(X_test, X_test),
                   verbose=0)
    return autoencoder
```

```python
encoder_upsampling, decoder_upsampling, autoencoder_upsampling = get_encoder_decoder_with_upsampling()
```

```python
encoder_conv_transpose, decoder_conv_transpose, autoencoder_conv_transpose = get_encoder_decoder_with_conv_transpose()
```

```python
autoencoder_conv_transpose  = compile_train(autoencoder_conv_transpose, X_train, X_test, 20)
```

```python
autoencoder_upsampling  = compile_train(autoencoder_upsampling, X_train, X_test, 20)
```

```python
encoded_imgs_conv_transpose = encoder_conv_transpose.predict(X_test)
decoded_imgs_conv_transpose = decoder_conv_transpose.predict(encoded_imgs_conv_transpose)
encoded_imgs_upsampling = encoder_upsampling.predict(X_test)
decoded_imgs_upsampling = decoder_upsampling.predict(encoded_imgs_upsampling)
```
